```python

## the object datatypes will be converted to numerical category in order to pass it to our algorithm
from sklearn import preprocessing as pp
```

```python
## instanciation
label_encoder=pp.LabelEncoder()
label_encoder
```

```python
## get the columns of the object datatypes as a list because i want to use for the loop iteration 
obj=Objects.columns.to_list()
obj
```

```python
for column in obj:
    Objects[column]=label_encoder.fit_transform(Objects[column])
```

```python
Objects
```

```python
df1=numeric.merge(Objects,right_index=True,left_index=True)
```

```python
df1
```

```python
X=df1.drop('occupational status',axis=1)
```

```python
X
```

```python
X=df1.drop(['State of origin','Time Taken you think the drug takes to work','is yes state the preferred antibiotics','Side effects of the preferred drugs','Class'],axis=1)
```

```python
y=df1['Class']
```

## spliting the data into training and testing dataset

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=123)
```

## import model to be used 

for this project im going to make use of four(4) models in machine learning to test and train the dataset

1. KNN
2. Random forest 
3. SVC 
4. logistic regression 

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import classification_report
from sklearn.metrics import roc_auc_score as roc
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
LOG_model=LogisticRegression()
LOG_model.fit(X_train,y_train)
y_pred_train_LOG=LOG_model.predict(X_train)
train_accuracy_LOG=accuracy_score(y_train,y_pred_train_LOG)
print('train_accuracy in LOG',train_accuracy_LOG)
print('==================================================')
y_pred=LOG_model.predict(X_test)
accuracy=accuracy_score(y_test,y_pred)
print('accuracy of LOG',accuracy)
print('==================================================')
```

```python
RFC_model=RandomForestClassifier()
RFC_model.fit(X_train,y_train)
##training the dataset
y_pred_train_RFC=RFC_model.predict(X_train)
train_accuracy_RFC=accuracy_score(y_train,y_pred_train_RFC)
print('train_accuracy in RFC',train_accuracy_RFC)
#print('==================================================')
## testing the model
y_pred_test_RFC=RFC_model.predict(X_test)
test_accuracy_RFC=accuracy_score(y_test,y_pred_test_RFC)
print('accuracy in RFC',test_accuracy_RFC)
print('===============================================')
```

```python
KNN_model=KNeighborsClassifier()
KNN_model.fit(X_train,y_train)
##training the dataset
y_pred_train_KNN=KNN_model.predict(X_train)
train_accuracy_KNN=accuracy_score(y_train,y_pred_train_KNN)
print('train_accuracy in KNN',train_accuracy_KNN)
print('==================================================')
## testing the model
y_pred_test_KNN=KNN_model.predict(X_test)
test_accuracy_KNN=accuracy_score(y_test,y_pred_test_KNN)
print('test accuracy in KNN',test_accuracy_KNN)
print('===============================================')
```

```python
## making using of Decision tree 
SVC_model=SVC()
SVC_model.fit(X_train,y_train)
##training the dataset
y_pred_train_SVC=SVC_model.predict(X_train)
train_accuracy_SVC=accuracy_score(y_train,y_pred_train_SVC)
print('train_accuracy in SVC',train_accuracy_SVC)
print('==================================================')
## testing the model
y_pred_test_SVC=SVC_model.predict(X_test)
test_accuracy_SVC=accuracy_score(y_test,y_pred_test_SVC)
print('accuracy in SVC',test_accuracy_SVC)
print('===============================================')
```
